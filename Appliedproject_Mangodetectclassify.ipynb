```python
import os
from ultralytics import YOLO
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# === CONFIGURATION ===
model_path = '/content/runs/detect/train/weights/best.pt'
image_dir = '/content/Mango-Leaf-Disease-6/test/images'
label_dir = '/content/Mango-Leaf-Disease-6/test/labels'

# === LOAD YOLOv8 MODEL ===
model = YOLO(model_path)

# === RUN INFERENCE ===
results = model.predict(source=image_dir, conf=0.25, iou=0.5, save=False, verbose=False)

true_labels = []
pred_labels = []

# === PROCESS RESULTS ===
for result in results:
    image_name = os.path.basename(result.path)
    label_path = os.path.join(label_dir, image_name.replace('.jpg', '.txt'))

    # ---- TRUE LABEL: First label only ----
    if os.path.exists(label_path):
        with open(label_path, 'r') as f:
            lines = f.readlines()
            if lines:
                true_cls = int(lines[0].strip().split()[0])
                true_labels.append(true_cls)
            else:
                continue  # No ground truth

    # ---- PREDICTED LABEL: Highest confidence only ----
    if result.boxes and len(result.boxes) > 0:
        best_box = result.boxes[0]
        pred_cls = int(best_box.cls[0])
        pred_labels.append(pred_cls)
    else:
        pred_labels.append(-1)  # No prediction

# === CHECK LENGTH ===
print(f" Matched {len(true_labels)} true vs {len(pred_labels)} predicted labels.")

# === CLASSIFICATION REPORT ===
print("\n Classification Report:")
print(classification_report(true_labels, pred_labels, digits=3))

# === CONFUSION MATRIX ===
cm = confusion_matrix(true_labels, pred_labels)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Class')
plt.ylabel('True Class')
plt.title('YOLOv8 Confusion Matrix (Top-1)')
plt.show()
```

```python
import os
from ultralytics import YOLO
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

# === CONFIGURATION ===
model_path = '/content/runs/detect/train/weights/best.pt'  # Path to YOLOv8 trained model
image_dir = '/content/Mango-Leaf-Disease-6/test/images'                         # Path to test images
label_dir = '/content/Mango-Leaf-Disease-6/test/labels'                         # Path to test labels in YOLO format (.txt files)

# === LOAD MODEL ===
model = YOLO(model_path)

# === RUN INFERENCE ===
results = model.predict(source=image_dir, conf=0.25, iou=0.5, save=False, verbose=False)

# === PREPARE LABELS ===
true_labels = []
pred_labels = []

# === LOOP OVER PREDICTIONS ===
for result in results:
    image_name = os.path.basename(result.path)
    label_file = os.path.join(label_dir, image_name.replace('.jpg', '.txt'))

    # ---- TRUE LABEL (first object only) ----
    if os.path.exists(label_file):
        with open(label_file, 'r') as f:
            lines = f.readlines()
            if lines:
                cls_true = int(lines[0].split()[0])
                true_labels.append(cls_true)
            else:
                continue  # skip if no ground truth

    # ---- PREDICTED LABEL (highest confidence) ----
    if result.boxes and len(result.boxes) > 0:
        top_box = result.boxes[0]  # highest confidence box
        cls_pred = int(top_box.cls[0])
        pred_labels.append(cls_pred)
    else:
        pred_labels.append(-1)  # if no prediction, use -1

# === ENSURE LENGTHS MATCH ===
if len(true_labels) != len(pred_labels):
    min_len = min(len(true_labels), len(pred_labels))
    true_labels = true_labels[:min_len]
    pred_labels = pred_labels[:min_len]

# === METRICS ===
print(f"\n Total Evaluated Images: {len(true_labels)}")
test_accuracy = accuracy_score(true_labels, pred_labels)
print(f" Test Accuracy: {test_accuracy * 100:.2f}%")

print("\n Classification Report:")
print(classification_report(true_labels, pred_labels, digits=3))

# === CONFUSION MATRIX ===
cm = confusion_matrix(true_labels, pred_labels)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("YOLOv8 Confusion Matrix (One-Label Per Image)")
plt.xlabel("Predicted Class")
plt.ylabel("True Class")
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, datasets, transforms
from torch.utils.data import DataLoader
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from torchvision.models import ResNet50_Weights

# Define device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# Load datasets
train_dataset = datasets.ImageFolder('/content/Mango leaf disease/train', transform=transform)
val_dataset   = datasets.ImageFolder('/content/Mango leaf disease/val', transform=transform)
test_dataset  = datasets.ImageFolder('/content/Mango leaf disease/test', transform=transform)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader  = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Load pretrained ResNet50
#model = models.resnet50(pretrained=True)
model = models.resnet50(weights=ResNet50_Weights.IMAGENET1K_V1)

# Modify final layer for your number of classes
num_classes = len(train_dataset.classes)
model.fc = nn.Linear(model.fc.in_features, num_classes)
model = model.to(device)

# Loss and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.0001)

# Class names
class_names = train_dataset.classes

# Tracking variables
num_epochs = 5
train_accuracy, val_accuracy = [], []
train_loss, val_loss = [], []

# Training and validation loop
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        running_loss += loss.item()

    train_accuracy.append(100 * correct / total)
    train_loss.append(running_loss / len(train_loader))
    print(f"Epoch [{epoch+1}/{num_epochs}] - Train Loss: {train_loss[-1]:.4f} - Accuracy: {train_accuracy[-1]:.2f}%")

    # Validation
    model.eval()
    val_correct = 0
    val_total = 0
    val_running_loss = 0.0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()
            val_running_loss += loss.item()

    val_accuracy.append(100 * val_correct / val_total)
    val_loss.append(val_running_loss / len(val_loader))
    print(f"Validation Accuracy: {val_accuracy[-1]:.2f}% - Val Loss: {val_loss[-1]:.4f}")

# Plot accuracy and loss
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(range(1, num_epochs + 1), train_accuracy, label='Train Accuracy')
plt.plot(range(1, num_epochs + 1), val_accuracy, label='Validation Accuracy')
plt.xlabel('Epochs'); plt.ylabel('Accuracy (%)'); plt.title('Accuracy'); plt.legend()

plt.subplot(1, 2, 2)
plt.plot(range(1, num_epochs + 1), train_loss, label='Train Loss')
plt.plot(range(1, num_epochs + 1), val_loss, label='Validation Loss')
plt.xlabel('Epochs'); plt.ylabel('Loss'); plt.title('Loss'); plt.legend()
plt.tight_layout()
plt.show()

# Final test evaluation
model.eval()
y_true = []
y_pred = []

with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        y_true.extend(labels.cpu().numpy())
        y_pred.extend(predicted.cpu().numpy())

# Test accuracy
test_accuracy = 100 * np.sum(np.array(y_true) == np.array(y_pred)) / len(y_true)
print(f"Test Accuracy: {test_accuracy:.2f}%")

# Classification report
print("\nClassification Report:")
print(classification_report(y_true, y_pred, target_names=class_names, digits=4))

# Confusion matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()
```

```python
# Hybrid YOLOv8 + ResNet50 Mango Leaf Disease Classification

import torch
import torch.nn as nn
from torchvision import models, transforms, datasets
from torch.utils.data import DataLoader
from PIL import Image
import numpy as np
from ultralytics import YOLO
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd
import cv2
from google.colab.patches import cv2_imshow

# -------------------- Setup --------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Class labels
class_names = ['Anthracnose', 'Bacterial_Canker', 'Die_Back', 'Gall_Mid', 'healthy']

# Image transformation
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# -------------------- Step 1: Load YOLOv8 --------------------
yolo_model = YOLO('/content/runs/detect/train/weights/best.pt')  # Path to your trained YOLO model
yolo_model.eval()

# -------------------- Step 2: Load and Fine-Tune ResNet50 --------------------
resnet_model = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
num_features = resnet_model.fc.in_features
resnet_model.fc = nn.Sequential(
    nn.Linear(num_features, 512),
    nn.ReLU(),
    nn.Dropout(0.4),
    nn.Linear(512, 5)
)
resnet_model.to(device)

# Load data
train_data = datasets.ImageFolder('/content/Mango leaf disease/train', transform=transform)
val_data = datasets.ImageFolder('/content/Mango leaf disease/val', transform=transform)
train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
val_loader = DataLoader(val_data, batch_size=32, shuffle=False)

# Training setup
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(resnet_model.parameters(), lr=0.0001)
scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)
num_epochs = 5
train_accuracy, val_accuracy = [], []

# Fine-tuning loop
for epoch in range(num_epochs):
    resnet_model.train()
    correct, total, running_loss = 0, 0, 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = resnet_model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        running_loss += loss.item()

    acc = 100 * correct / total
    train_accuracy.append(acc)
    print(f"Epoch {epoch+1}, Loss: {running_loss:.4f}, Train Accuracy: {acc:.2f}%")

    # Validation
    resnet_model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for inputs, labels in val_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = resnet_model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    acc = 100 * correct / total
    val_accuracy.append(acc)
    print(f"Validation Accuracy: {acc:.2f}%")
    scheduler.step()

# Save model
torch.save(resnet_model.state_dict(), 'resnet50_finetuned.pth')

# Accuracy plot
plt.plot(range(1, num_epochs+1), train_accuracy, label='Train')
plt.plot(range(1, num_epochs+1), val_accuracy, label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training vs Validation Accuracy')
plt.legend()
plt.show()

# -------------------- Step 3: Hybrid Prediction --------------------
def hybrid_predict(image_path, conf_thresh=0.5):
    image = Image.open(image_path).convert("RGB")
    results = yolo_model(image_path)[0]
    predictions = []
    for box in results.boxes:
        if box.conf.item() < conf_thresh:
            continue
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
        crop = image.crop((x1, y1, x2, y2))
        crop_tensor = transform(crop).unsqueeze(0).to(device)
        with torch.no_grad():
            output = resnet_model(crop_tensor)
            _, predicted = torch.max(output, 1)
            predictions.append(predicted.item())
    return predictions

# -------------------- Step 4: Evaluation --------------------
y_true, y_pred = [], []
test_root = '/content/Mango leaf disease/test'
for cls_idx, cls in enumerate(class_names):
    folder_path = os.path.join(test_root, cls)
    if not os.path.exists(folder_path):
        continue
    for img_file in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_file)
        preds = hybrid_predict(img_path)
        for p in preds:
            y_true.append(cls_idx)
            y_pred.append(p)

# Classification Report
report = classification_report(y_true, y_pred, target_names=class_names, output_dict=True)
report_df = pd.DataFrame(report).transpose()
print(report_df.round(2))

# Save report
report_df.to_csv("classification_report.csv")

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()

# Test accuracy
test_acc = accuracy_score(y_true, y_pred)
print(f"Final Test Accuracy: {test_acc*100:.2f}%")

# -------------------- Step 5: Visualize Predictions --------------------
output_dir = '/content/Mango-Leaf-Disease-6/Out'
os.makedirs(output_dir, exist_ok=True)
image_extensions = ['.jpg', '.jpeg', '.png']

image_paths = []
for root, dirs, files in os.walk(test_root):
    for file in files:
        if os.path.splitext(file)[1].lower() in image_extensions:
            image_paths.append(os.path.join(root, file))

print(f"Visualizing {len(image_paths)} images...")
for image_path in image_paths:
    img = cv2.imread(image_path)
    if img is None:
        continue
    results = yolo_model(image_path)[0]
    for box in results.boxes.data.tolist():
        x1, y1, x2, y2, conf, class_id = box
        x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
        if int(class_id) == 4:
            label = 'healthy'
        else:
            crop = img[y1:y2, x1:x2]
            if crop.size == 0: continue
            crop_pil = Image.fromarray(crop)
            input_tensor = transform(crop_pil).unsqueeze(0).to(device)
            with torch.no_grad():
                output = resnet_model(input_tensor)
                _, pred = torch.max(output, 1)
                label = class_names[pred.item()]
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, label, (x1, y1-10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)
    out_path = os.path.join(output_dir, os.path.basename(image_path))
    cv2.imwrite(out_path, img)
    cv2_imshow(img)

print(" Hybrid inference and visualization complete.")
```

```python
import os
import cv2
import csv
import torch
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from PIL import Image
from ultralytics import YOLO
from torchvision import models, transforms
from sklearn.metrics import classification_report, confusion_matrix
from google.colab.patches import cv2_imshow
from torchvision.models import ResNet50_Weights

# -------------------- Setup --------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
class_names = ['Anthracnose', 'Bacterial_Canker', 'Die_Back', 'Gall_Mid', 'healthy']

# -------------------- Load YOLOv8 --------------------
yolo_model = YOLO('/content/runs/detect/train/weights/best.pt')  # fine-tuned YOLO
yolo_model.eval()

# -------------------- Load ResNet50 --------------------
resnet_model = models.resnet50(weights=ResNet50_Weights.IMAGENET1K_V1)
num_ftrs = resnet_model.fc.in_features
resnet_model.fc = torch.nn.Sequential(
    torch.nn.Linear(num_ftrs, 512),
    torch.nn.ReLU(),
    torch.nn.Dropout(0.4),
    torch.nn.Linear(512, 5)
)
resnet_model.load_state_dict(torch.load('/content/resnet50_finetuned.pth', map_location=device))
resnet_model.to(device)
resnet_model.eval()

# -------------------- Transform --------------------
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

# -------------------- Directories --------------------
test_root = '/content/Mango leaf disease/test'
output_dir = '/content/Mango-Leaf-Disease-6/Out'
os.makedirs(output_dir, exist_ok=True)
image_extensions = ['.jpg', '.jpeg', '.png']

# -------------------- CSV Results Setup --------------------
csv_path = '/content/hybrid_inference_results.csv'
csv_file = open(csv_path, mode='w', newline='')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['Image_Name', 'X1', 'Y1', 'X2', 'Y2', 'Confidence', 'Predicted_Class', 'True_Class'])

# -------------------- Hybrid Inference --------------------
y_true = []
y_pred = []

image_paths = []
for root, _, files in os.walk(test_root):
    for file in files:
        if os.path.splitext(file)[1].lower() in image_extensions:
            image_paths.append(os.path.join(root, file))

print(f" Found {len(image_paths)} images")

for image_path in image_paths:
    img = cv2.imread(image_path)
    if img is None:
        continue

    # Extract true class from folder name
    true_class = os.path.basename(os.path.dirname(image_path))
    true_class_idx = class_names.index(true_class) if true_class in class_names else -1

    results = yolo_model(image_path)[0]
    for box in results.boxes.data.tolist():
        x1, y1, x2, y2, conf, class_id = box
        x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])
        label = class_names[int(class_id)] if int(class_id) == 4 else 'unknown'

        # Classify with ResNet if not already healthy
        if int(class_id) != 4:
            crop = img[y1:y2, x1:x2]
            if crop.size == 0:
                continue
            crop_pil = Image.fromarray(crop)
            input_tensor = transform(crop_pil).unsqueeze(0).to(device)
            with torch.no_grad():
                output = resnet_model(input_tensor)
                _, pred = torch.max(output, 1)
                pred_class = class_names[pred.item()]
        else:
            pred_class = 'healthy'

        # Add to metrics
        if true_class_idx != -1:
            y_true.append(true_class_idx)
            y_pred.append(class_names.index(pred_class))

        # Draw bounding box and label
        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, pred_class, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

        # Save to CSV
        csv_writer.writerow([
            os.path.basename(image_path), x1, y1, x2, y2, round(conf, 2), pred_class, true_class
        ])

    out_path = os.path.join(output_dir, os.path.basename(image_path))
    cv2.imwrite(out_path, img)
    cv2_imshow(img)

csv_file.close()
print(f" Inference done. Results saved to: {csv_path}")

# -------------------- Evaluation --------------------
print("\n Classification Report:")
report = classification_report(y_true, y_pred, target_names=class_names, digits=2)
print(report)

print("\n Confusion Matrix:")
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.show()
```
